```python
import pandas as pd
```

```python
movie = pd.read_csv("Resources/tmdb_5000_movies.csv")
movie.head()
crew = pd.read_csv("Resources/tmdb_5000_credits.csv")
crew.head()
```

```python
director = crew.crew.str.split("Directing", expand = True).filter([1])
newfilters = director.rename(columns = {1: 'director'})
newfilters = newfilters.director.str.split(':',expand = True).filter([4]).rename(columns = {4:'director'})
movie = newfilters.director.str.split('}',expand = True).filter([0]).rename(columns = {0:'director'}).merge(movie, right_index = True, left_index = True)
movie.head()
```

```python
filters = movie.production_companies.str.split(',',expand = True).filter([0])
newfilters = filters.rename(columns = {0: 'Company'})
newfilters = newfilters.Company.str.split(':',expand = True).filter([1])
movie = newfilters.rename(columns = {1:"production_company"}).merge(movie, right_index = True, left_index = True)
movie

```

```python
genre = movie.genres.str.split(',',expand = True).filter([1])
newgenre= genre.rename(columns = {1: 'genre'})
newgenre = newgenre.genre.str.split(':',expand = True).filter([1])
newgenre= newgenre.rename(columns = {1: 'genre'})
newgenre = newgenre.genre.str.split('}',expand = True).filter([0])
movie= newgenre.rename(columns = {0: 'genre'}).merge(movie, right_index = True, left_index = True)
movie
```

```python
movie = movie.drop('genres',1)
movie = movie.drop('production_companies',1)
movie = movie.drop('homepage',1)
movie = movie.drop('keywords',1)
movie = movie.drop('original_language',1)
movie = movie.drop('spoken_languages',1)
movie = movie.drop('production_countries',1)
movie = movie.drop('original_title',1)
```

```python
movie=movie.rename(columns = {'id':'identification'})
movie.head()
```

```python
movie.to_csv('Resources/cleanedmovies.csv')
```
